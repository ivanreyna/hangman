```python
import random
from words import word_list
# Get word from list with specific length
def get_word(length):
    l_words = [word for word in word_list if len(word) == length]
    if l_words:
        return random.choice(l_words).upper()
    else:
        return None
# Function to play
def run(word):
    complete_word = "_" * len(word)  # Show guessed letters
    guessed = False  # Indicates if word is guessed
    guessed_letters = []  # Store guessed letters
    guessed_words = []  # Store guessed words
    attempts = 6  # attempts done by user (defined by inputs)
    score = 0
    score = max(score, 0)  # just to avoid negatives
    print("The Hangman Game done by @ivanreyna!")
    print(hangman(attempts))
    print(complete_word)
    print("\n")
    
    while not guessed and attempts > 0:
        guess = input("Please guess a letter or word: ").upper()

        # If the guess is a single letter
        if len(guess) == 1 and guess.isalpha():
            if guess in guessed_letters:
                print("You already guessed the letter", guess)
            elif guess not in word:
                print(guess, "is not in the word.")
                attempts -= 1
                score -= 5  # Substract points for incorrect guess
                guessed_letters.append(guess)
            else:
                print("Good job,", guess, "is in the word!")
                guessed_letters.append(guess)
                # Reveal the guessed letter in the word
                word_as_list = list(complete_word)
                indices = [word for word, letter in enumerate(word) if letter == guess] #enumerate() to iterate instead of doinf for i in range len(word)
                for index in indices:
                    word_as_list[index] = guess
                complete_word = "".join(word_as_list)
                score += 10  # Add points for correct guess
                if "_" not in complete_word:
                    guessed = True
        
        # If the guess is a word
        elif len(guess) == len(word) and guess.isalpha():
            if guess in guessed_words:
                print("You already guessed the word", guess)
            elif guess != word:
                print(guess, "is not the word.")
                attempts -= 1
                score -= 5  # Deduct points for incorrect guess
                guessed_words.append(guess)
            else:
                guessed = True
                complete_word = word
                score += 10 * len(word)  # Bonus for guessing the whole word
        
        else:
            print("Not a valid guess.")
        
        print(f"Score: {score}")
        print(complete_word)
        print(hangman(attempts))
        print("\n")

    if guessed:
        print("You win!")
    else:
        print("Sorry, too many attempts. The word was " + word + ". Try again next time!")
    print(f"Final Score: {score}")
# hangman
def hangman(attempts):
    dies = [  # final state: head, torso, both arms, and both legs
                """
                   --------
                   |      |
                   |      O
                   |     \|/
                   |      |
                   |     / \
                   -
                """,
                """
                   --------
                   |      |
                   |      O
                   |     \|/
                   |      |
                   |     / 
                   -
                """,
                """
                   --------
                   |      |
                   |      O
                   |     \|/
                   |      |
                   |      
                   -
                """,
                """
                   --------
                   |      |
                   |      O
                   |     \|
                   |      |
                   |     
                   -
                """,
                """
                   --------
                   |      |
                   |      O
                   |      |
                   |      |
                   |     
                   -
                """,
                """
                   --------
                   |      |
                   |      O
                   |    
                   |      
                   |     
                   -
                """,
                """
                   --------
                   |      |
                   |      
                   |    
                   |      
                   |     
                   -
                """
    ]
    return dies[attempts]

# Main function to start the game
def main():
    while True:
        try:
            length = int(input("Enter the desired word length: "))
            word = get_word(length)
            if word:
                run(word)
            else:
                print(f"No words of length {length} found. Please try again.")
        except ValueError:
            print("Please enter a valid number.")
        if input("Play Again? (Y/N) ").upper() != "Y":
            break

if __name__ == "__main__":
    main()
```
